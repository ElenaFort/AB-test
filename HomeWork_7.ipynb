# Урок 7. Нестандартные ситуации
На основе выбранных продуктов из первого вебинара, выберите 2 A/B-теста и составьте план эксперимента, чтобы избежать Peeking Problem

Распишите, какие метрики и как вы будете анализировать

## Эксперимент 1

### Гипотеза:
Если в СберМаркет добавить больше магазинов - партнёров, то количество заказов увеличится на Х %, потомучто мы таким образом расширим ассортимент, а значит клиент сможет найти то, что ему нужно с большей вероятностью.

### Что делаем:
**Контрольная версия:** оставляем приложение без изменений

**Тестовая версия:** добавляем один магазин - партнёр

### На каких пользователях тестируем:

На 10 % всех пользователей приложения

### Ключевые метрики:

- количество покупок, сделанных пользователями
- количество повторных заказов

### Ожидаемый эффект:

Увеличение количества заказов

### План действий:

- Если результат эксперимента будет положительным и прирост значений ключевых метрик будет не менее ожидаемого, то распространяем изменения на всех пользователей.

- Если значения ключевых метрик снижаются или имеют незначительный прирост, то откатываем эксперимент.

Чтобы избежать Peeking Problem, необходимо придерживаться расчетов, выполненных перед началом эксперимента (время и количество трафика). Принимать решения на основе неполных данных нельзя. Ждем окончания эксперимента и потом принимаем решение.
