### **СберМаркет**

1. *Эффективность промо акций*

 Позволяет определить оптимальную скидку — на сколько % увеличится спрос при увеличении скидки на X% при разных факторах.

2. *Средний чек и количество товаров*

 Позволяет отслеживать популярность тех или иных продуктов в различные сезоны. Благодаря этому можно сконцентрироваться на продвижении конкретных категорий или продуктов для стимулирования продаж. Анализ среднего чека позволяет компании не только держать руку на пульсе, но и прогнозировать доходы и управлять продажами.

3. *Время сборки и доставки заказов*

 Клиентам важно получать заказы быстро и в точно определенный срок.

4. *Количество новых и повторных клиентов*

 Наилучшим показателем того, на сколько довольны клиенты является тот факт, продолжают ли они с вами сотрудничать. Когда у них снова возникает необходимость приобрести продукт, остаются ли они с вами или переходят к конкуренту?

 Успешная работа с клиентами увеличивает как количество клиентов, так и процент тех клиентов, которые совершают повторные покупки.

5. *Рейтинг сборщиков и курьеров*

 Чем более довольны покупатели сервисом, тем больше вероятность повторных заказов
